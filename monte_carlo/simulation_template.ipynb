### Analyze the results


The following is for an average. THIS IS NOT NECESSARY IF YOU DON'T NEED AN AVERAGE


#### Calculate the sum of all trials in the list


```python
# Create variables
total_trials = len(consecutive_days_results)
sum_of_trials = 0

# Loop through each entry, adding it to the sum
for i in consecutive_days_results:
    sum_of_trials = sum_of_trials + i
        
# Print the results
print(sum_of_trials)
```

#### Calculate the average

- Average = sum of trials / total trials


```python
# Calculate the average
average = sum_of_trials/total_trials

# Print the results
print(average)
```

### Graph the results

## Histogram


```python
import matplotlib.pyplot as plt #importing a module for plotting
%matplotlib inline 
# this allows us to show graphs in a notebook

# Create the bins array
bins = []
length = len(count_consecutive_days)
for i in range(length):
    bins.append(i)
print(bins)

# plt.hist(array, dividers/bins, alignment, graph color, border color)
plt.hist(consecutive_days_results,bins,align='mid',color="blue",edgecolor='k')

# Labels
plt.title("Frequency of Consecutive days before late once")
plt.ylabel("Frequency")
plt.xlabel("Days before late")

plt.show()
```

### "Grouped" Histogram


```python
# Import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline

# Create bins array in groups of 5
bins = []
for i in range(12):
    bin = 5*i
    bins.append(bin)

# Create title and lables
plt.title("Frequency of consecutive days on time before late")
plt.ylabel("Frequency")
plt.xlabel("Consecutive")

# plt.hist(array, dividers/bins, alignment, graph color, border color)
plt.hist(consecutive_days_results, bins, align="mid", color="lightgreen", edgecolor="black")

# Show
plt.show()
```
