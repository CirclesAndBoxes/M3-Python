```python
import numpy as np
import numpy.linalg as la
import math

# g'(t) = a x1(t) + b x2(t)
# h'(t) = c x1(t) + d x2(t)
# Plug in for a, b, c, d

# given a = -1, b = -0.1, c = 0.6, d = -0.2
transition_matrix = np.array([[-1, -.1],
                              [.6, -.2]])


eigvals, eigvecs = la.eig(transition_matrix)

# Note eigvecs is stored as a 2D array where eigenvectors

initial_condition = np.array([100,0])

[c1, c2] = tuple(la.solve(eigvecs, initial_condition))

def x1(t):
    return (eigvecs[:,0] * c1 * math.e ** (eigvals[0] * t) 
        + eigvecs[:,1] * c2 * math.e ** (eigvals[1] * t))[0]

def x2(t):
    return (eigvecs[:,0] * c1 * math.e ** (eigvals[0] * t) 
        + eigvecs[:,1] * c2 * math.e ** (eigvals[1] * t))[1]

# uncomment below to plot x1(x) and x2(x)
# plot_functions(c1, c2, eigvals, eigvecs)

print(x1(1), x2(2))

hours = 4
```

```python
import numpy as np
import matplotlib.pyplot as plt

def plot_functions(f, g, x_range=(-10, 10), num_points=1000):
    """
    Plots two functions f(x) and g(x) on the same graph.
    
    Parameters:
    - f: Function 1 (callable)
    - g: Function 2 (callable)
    - x_range: Tuple (min, max) specifying the x-axis range
    - num_points: Number of points to plot
    """
    x = np.linspace(x_range[0], x_range[1], num_points)
    # vectorize is an INSANE function. Figure out how to use.
    y_f = np.vectorize(f)(x)
    y_g = np.vectorize(g)(x)
    
    plt.figure(figsize=(8, 5))
    plt.plot(x, y_f, label='first(x)', color='blue')
    plt.plot(x, y_g, label='second(x)', color='red', linestyle='dashed')
    
    plt.axhline(0, color='black', linewidth=0.5)
    plt.axvline(0, color='black', linewidth=0.5)
    plt.legend()
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title('Plot of first(x) and second(x)')
    plt.grid(True, linestyle='--', linewidth=0.5)
    plt.show()

# Example usage:
# plot_functions(lambda x: x**2, lambda x: np.sin(x))
```
